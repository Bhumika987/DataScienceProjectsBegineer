#  Label Encoder
In ML, we usually ddeal with dataset which contains multiple labels in one or more than one columns. These labels can be in the form of words or numbers. Label Encoding refers to converting the labels into numeric form so as to convert it into the machine-readable format.


```python
#converting the dataset in numeric form
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
df.head()
```

# Model Training


```python
# training and testing 
from sklearn.model_selection import train_test_split
#train - 70
#test - 30
X =df.drop(columns=['Species'])
Y =df['Species']
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.30)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
#print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# knn - k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```
